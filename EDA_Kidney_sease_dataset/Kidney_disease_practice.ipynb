```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')
```

```python
df.columns = ['id','age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell', 'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium','potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count','hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema','aanemia', 'class']
```

```python
Categorical_column = list(df.dtypes[df.dtypes=='O'].index)
Numerical_column = list(df.dtypes[df.dtypes!='O'].index)
```

```python
Categorical_column
Numerical_column
```

```python
for col in Categorical_column:
    print(f' The {col} have unique feature :--- {df[col].unique()}')
```

```python
col_to_remove = ['packed_cell_volume','white_blood_cell_count','red_blood_cell_count','hypertension','diabetes_mellitus','coronary_artery_disease','class',]
char_to_remove = ['\t','\t?']
```

```python
for char in char_to_remove:
    for col in col_to_remove:
        df[col]=df[col].str.replace(char,'')
```

```python
column_to_numeric = ['packed_cell_volume','white_blood_cell_count','red_blood_cell_count']
for col in column_to_numeric:
    df[col] =pd.to_numeric(df[col])
```
