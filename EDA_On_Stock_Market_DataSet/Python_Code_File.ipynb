```python
reason = ['Reason_Equity','Reason_Mutual','Reason_Bonds','Reason_FD']
```

```python
df[reason]
```

```python
for col in reason:
    print(df[col].value_counts())
```

```python
# Insight >>
# For investment in Equity >> most persion invest because Reason is capital appreaciation
# For Investment in Mutual >> It's give better returns and for diversity the money is the reson for investment in Mutual Fund
# Reason for Bond >> It's give Assurity of money and second reason is Safe investment
#  Reason of FD >> It's Risk Free investment and Fixed Returns are reason for taking most investment
```

```python
for col in reason:
    plt.figure(figsize = (12,8))
    sns.countplot(x=col,data = df, palette='viridis')
    plt.title(f'Investment in {col}')
    plt.show()
```
