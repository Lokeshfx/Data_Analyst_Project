```python
#1 MB 1024 kb
def size_process(item):
    if str(item)[-1] == 'M':
        res = float(str(item).replace('M', ''))
        res = res*1024
        return res
    elif str(item)[-1] == 'k':
        res = float(str(item).replace('k', ''))
        return res
    else:
        return ""
```

```python
df['Size'] = df['Size'].apply(size_process)
```
